```python
spreadsheetId = '1Ycg7zTxds9DZnDvTrFcyNNKuTUxg6Yy6WF0a8Wc02WQ'
range_name = 'managers!A:C'
table = service.spreadsheets().values().get(spreadsheetId=spreadsheetId, range=range_name).execute()
```

```python
range_name = 'clients!A:C'
table = service.spreadsheets().values().get(spreadsheetId=spreadsheetId, range=range_name).execute()
aa = []
bb = []
cc= []
for i in table['values'][1:]:
    if len(i)==1: 
        aa.append( i[0] )
        bb.append( None )
        cc.append( None )
    else:
        aa.append( i[0] )
        bb.append( i[1] )
        cc.append( i[2] )

data_clients = pd.DataFrame( list(zip(aa, bb, cc) ),columns=[ 
                                                table['values'][0][0],
                                                table['values'][0][1],
                                                table['values'][0][2]])
data_clients.head(5)
```

```python
range_name = 'leads!A:F'
table = service.spreadsheets().values().get(spreadsheetId=spreadsheetId, range=range_name).execute()

aa = []
bb = []
cc= []
dd=[]
ee=[]
ff=[]
for i in table['values'][1:]:
    aa.append( i[0] )
    bb.append( i[1] )
    cc.append( i[2] )
    dd.append( i[3] )
    ee.append( i[4] )
    ff.append( i[5] )

data_leads_all = pd.DataFrame( list(zip(aa, bb, cc, dd, ee, ff) ),columns=[ 
                                                table['values'][0][0],table['values'][0][1],
                                                table['values'][0][2],table['values'][0][3],
                                                table['values'][0][4],table['values'][0][5]])
data_leads_all.head(5)
```

```python
range_name = 'transactions!A:D'
table = service.spreadsheets().values().get(spreadsheetId=spreadsheetId, range=range_name).execute()

aa = []
bb = []
cc= []
dd=[]

for i in table['values'][1:]:
    aa.append( i[0] )
    bb.append( i[1] )
    cc.append( i[2] )
    dd.append( i[3] )


data_tran = pd.DataFrame( list(zip(aa, bb, cc, dd) ),columns=[ 
                                                table['values'][0][0],table['values'][0][1],
                                                table['values'][0][2],table['values'][0][3]
])
data_tran.head(5)
```

```python
data_leads_all['created_at']=pd.to_datetime(data_leads_all['created_at'])
data_tran['created_at']=pd.to_datetime(data_tran['created_at'])
data_tran['m_real_amount']=pd.to_numeric(data_tran['m_real_amount'])

data_clients=data_clients.sort_values('created_at')
data_tran=data_tran.sort_values('created_at')
data_leads_all=data_leads_all.sort_values('created_at')
```

```python
StartReport=datetime.datetime(2020, 11, 1)
EndReport=datetime.datetime(2020, 12, 1)

data_leads_mgr=pd.merge(data_leads_all, data_mgr, left_on='l_manager_id', right_on='manager_id', how='left')
data_leads_mgr['d_club'].fillna('NotEnter', inplace=True)
data_leads=data_leads_mgr.loc[(data_leads_mgr['created_at']>=StartReport)&(data_leads_mgr['created_at']<=EndReport)].copy()


data_leads['NewLead']=0
data_leads['isBuy']=0
data_leads['SumAmount']=0
data_leads['NewBuy']=0
data_leads.astype({'SumAmount':int})

data_leads['d_utm_source'].fillna('NotEnter', inplace=True)

data_leads['Trash']=[1 if x=='00000000-0000-0000-0000-000000000000' else 0 for x in data_leads['l_client_id']]
data_leads['d_utm_source']=["NotEnter" if x=='' else x for x in data_leads['d_utm_source']]

```

```python
for i,ds in data_leads.iterrows():
    MinLeadDate=data_leads_all.loc[data_leads_all['l_client_id']==ds['l_client_id']]['created_at'].min()
    MinTranDate=data_tran.loc[data_tran['l_client_id']==ds['l_client_id']]['created_at'].min()
    MinDate=min(MinLeadDate, MinTranDate)
    
    if (ds['Trash']==0)&(ds['created_at']==MinDate):
        data_leads.loc[i,'NewLead']=1
    else: 
        data_leads.loc[i,'NewLead']=0
    
    tran_count=data_tran.loc[
        (data_tran['l_client_id']==ds['l_client_id'])&
        (data_tran['created_at']>=ds['created_at'])&
        (data_tran['created_at']<=ds['created_at']+datetime.timedelta(days=7))
        ]['transaction_id'].count()
    if (tran_count>0)&(ds['Trash']==0):
        data_leads.loc[i,'isBuy']=1
    else:
        data_leads.loc[i,'isBuy']=0

    if (ds['Trash']==0)&(ds['created_at']==MinDate)&(tran_count>0):
        data_leads.loc[i, 'NewBuy']=1
    else:
        data_leads.loc[i, 'NewBuy']=0

for i, ds in data_leads.loc[data_leads['NewBuy']==1].iterrows():
    data_leads.loc[i,'SumAmount']=data_tran.loc[
        (data_tran['created_at']>=ds['created_at'])&
        (data_tran['created_at']<=EndReport)&
        (data_tran['l_client_id']==ds['l_client_id'])
    ]['m_real_amount'].sum()
    

data_leads['SumAmount']=pd.to_numeric(data_leads['SumAmount'])
```

```python
pt_Total=pd.pivot_table(data_leads, index=["d_utm_source","d_club","l_manager_id"], 
               values=["lead_id","NewLead", "isBuy","Trash","NewBuy","SumAmount"], 
               aggfunc={'lead_id':len, 'NewLead':np.sum, 'isBuy':np.sum, 'Trash':np.sum, 'NewBuy':np.sum, 'SumAmount':np.sum})

pt_Total=pt_Total.rename_axis(None, axis=1).reset_index()
```
